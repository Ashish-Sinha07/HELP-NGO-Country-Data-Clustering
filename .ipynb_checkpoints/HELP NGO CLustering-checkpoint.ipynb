# Data Preprocessing

```python
#Check for missing value
df.isnull().sum().sum()

# No missing values
```

```python
# Lets treat the extreme outliers
# we will go with capping using IQR method
# In order to cap only extreme ones lets keep whisker length = 2.5

for i in df.columns:
    Q1,Q3 = np.quantile(df[i],[0.25,0.75])
    IQR = Q3 - Q1
    
    ul,ll = Q3+2.5*IQR,Q1-2.5*IQR
    df[i] = df[i].apply(lambda x: ul if x>ul else ll if x<ll else x)

```

```python
t=1
plt.figure(figsize=(10,7))
for i in df.columns:

    plt.subplot(3,3,t)
    sns.boxplot(data=df,x=i,color='lightblue')
    plt.title(f'Skewness:{round(df[i].skew(),2)}')
    t+=1
plt.tight_layout()
plt.show()
```

```python
# Lets apply PCA on the data defore we procced with clustering. Just to have better visualization and for better clustering

# IN PCA we need to standarized data. SO lets import function from libraries

# First lets crate a instance of scaler
scaler  = StandardScaler()

# Also lets create a instance of PCA
pca = PCA(n_components=5,random_state=2)

# Lets create pipeline for above two processes
pipeline =Pipeline([('Scale',scaler),
                    ('PCA',pca)])
```

```python
pipeline.fit(df)
```

```python
pca.explained_variance_ratio_.cumsum()

# At n_components = 5, we able to explain approxametly ~95% variance, which is good enough.
# So, 9 dimensions data reeduce to 5 dimenstions
```

```python
final_data = pipeline.transform(df)
```

```python
#tranform return array. Lets convert it  in dataframe
final_data = pd.DataFrame(final_data,columns=['PC1','PC2','PC3','PC4','PC5'],index=df.index)
```

```python
final_data
```

# K-Means Clustering

```python
# Identify the value of K

# KElbow Visualizer

kelbow = KElbowVisualizer(estimator=KMeans(random_state=2),k=10)
kelbow.fit(final_data)
kelbow.show()
```

Although the above plot marks the elbow at k=6,. But we can see that significant drop in the distortion score from k=2 to k-3. As we increase the number of cluster it might become over fitted. Lets expore sillhouette visualizer to identify the correct or more generalize value of k.

```python
t=1
plt.figure(figsize=(12,20))
for i in range(2,11):
    plt.subplot(5,2,t)
    model = KMeans(n_clusters=i,random_state=2)
    sil_viz = SilhouetteVisualizer(model)
    sil_viz.fit(final_data)
    plt.title(f'K:{i} Sillhouette Score: {round(sil_viz.silhouette_score_,4)}')
    t+=1
plt.tight_layout()
plt.show()
```

After analyzing Elbow plot and Sillhoutte Visualizer lets go with k=3. Although there are few datapints that have bween misclassified, but we see that no drop in sillhoutte score and there is good enough data in every cluster. Also from k-elbow plot significant drop in WCSS (Distortion score) was observed from k=2 to k=3.

```python
final_kmeans = KMeans(n_clusters=3,random_state=2)
```

```python
labels = final_kmeans.fit_predict(final_data)
```

```python
# lets check sillhouette score as well
print('Sillhouette Score:',silhouette_score(final_data,labels))
```

```python
# Lets put these label on PCA Data
final_data['labels'] = labels
final_data.head()
```
